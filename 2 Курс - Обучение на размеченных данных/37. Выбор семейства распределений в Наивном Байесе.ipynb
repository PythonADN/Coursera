Один из самых известных алгоритмов машинного обучения, основной задачей которого является восстановление плотностей распределения данных обучающей выборки. Зачастую этот метод дает хорошее качество в задачах именно многоклассовой классификации. 

```python
from sklearn import datasets, cross_validation, metrics
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB # подключим модели Наивного Байеса
```

```python
# Загрузим данные из датасетов
Digits = datasets.load_digits() # первый датасет
Breast_cancer = datasets.load_breast_cancer() # второй датасет
```

```python
print "Ключи Digits: {}".format(Digits.keys())
print "Ключи Breast_cancer: {}".format(Breast_cancer.keys())
```

```python
from pandas import DataFrame

# создаем таблицы из дыннх по data
digits = DataFrame(Digits.data)
breast_cancer = DataFrame(Breast_cancer.data)
```

```python
digits.head(5)
```

```python
breast_cancer.head(5)
```

### Распределение Бернули

```python
# Первая модель Digits
model_1 = BernoulliNB()
model_1.fit(Digits.data, Digits.target)

# Построим кросс-валидацию для оценки модели
score  = cross_validation.cross_val_score(model_1, Digits.data, Digits.target)

# средняя оценка модели
print score.mean()

# ---------------------------

# Первая модель Breast_cancer
model_12 = BernoulliNB()
model_12.fit(Breast_cancer.data, Breast_cancer.target)

# Построим кросс-валидацию для оценки модели
score  = cross_validation.cross_val_score(model_12, Breast_cancer.data, Breast_cancer.target)

# средняя оценка модели
print score.mean()
```

### Мультиномиальное распределение

```python
# Вторая модель Digits
model_2 = MultinomialNB()
model_2.fit(Digits.data, Digits.target)

# Построим кросс-валидацию для оценки модели
score  = cross_validation.cross_val_score(model_2, Digits.data, Digits.target)

# средняя оценка модели
print score.mean()

# ---------------------------

# Вторая модель Breast_cancer
model_22 = MultinomialNB()
model_22.fit(Breast_cancer.data, Breast_cancer.target)

# Построим кросс-валидацию для оценки модели
score  = cross_validation.cross_val_score(model_22, Breast_cancer.data, Breast_cancer.target)

# средняя оценка модели
print score.mean()
```

### Гауссовское Распределение

```python
# Третья модель Digits
model_3 = GaussianNB()
model_3.fit(Digits.data, Digits.target)

# Построим кросс-валидацию для оценки модели
score  = cross_validation.cross_val_score(model_3, Digits.data, Digits.target)

# средняя оценка модели
print score.mean()

# ---------------------------

# Третья модель Breast_cancer
model_32 = GaussianNB()
model_32.fit(Breast_cancer.data, Breast_cancer.target)

# Построим кросс-валидацию для оценки модели
score  = cross_validation.cross_val_score(model_32, Breast_cancer.data, Breast_cancer.target)

# средняя оценка модели
print score.mean()
```
