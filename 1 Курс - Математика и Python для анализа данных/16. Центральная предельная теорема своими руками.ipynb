- Подкючаем необходимые библиотеки

```python
import numpy as np
import scipy.stats as sts
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import math
```

- Зададим $\lambda$ и число выборок для распределения Пуассона 
- Согласно Википедии, для распределения Пуассона математическое ожидание EX и дисперсия DX равны $\lambda$:
https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9F%D1%83%D0%B0%D1%81%D1%81%D0%BE%D0%BD%D0%B0

```python
l = 5 # зададим labda
p = sts.poisson(l) # зададаем распределенне Пуассона

M = 1000 # число выборок
```

- Пусть количество значений в выобрке $n=5$
- Построим гистаграмму распределения выборочных средних и теоретическую плотность нормального распределения

```python
n = 5 # количество случапйных значений в одной выборке
average = [] # создаем пустой массив средних случайных значений

# Определим M=1000 средних значений (average) по случайным выборкам (p.rvs),
# каждая из которых содержит n случайных значений из распределения Пуассона 
for i in range(1,M+1):
    average.append( np.sum(p.rvs(n))/float(n)  )

# построим Гистаграмму распределения выборочных средних (average) с числом карманов=30
plt.hist(average, 30, normed=True, facecolor='green', alpha=0.75, label='Poisson')

# на том же графике построим теоретическую плотность нормального распределения:
# для распределения Пуассона математическое ожидание и дисперсия EX=DX=lambda
x = np.linspace(2, 8, M)
pdf = sts.norm.pdf( x, l, math.sqrt(float(l)/n) )
plt.plot(x, pdf, alpha=0.75, lw=3, label='Theoretical pdf')

# создадим клетку
plt.grid(True)
# Обозначим Оси
plt.ylabel('$f(x_a)$')
plt.xlabel('$X_a$')
plt.legend(loc='upper left')

plt.show()
```
