## Постановка
### Влияние внешной привлекательности на уровень зароботной платы с учетом остальных факторов

По 1260 опрошенным имеются следующие данные:

* заработная плата за час работы, $; (wage)
* опыт работы, лет;
* образование, лет;
* внешняя привлекательность, в баллах от 1 до 5; (looks)
* бинарные признаки: пол, семейное положение, состояние здоровья (хорошее/плохое), членство в профсоюзе, цвет кожи (белый/чёрный), занятость в сфере обслуживания (да/нет).

Требуется оценить влияние внешней привлекательности на уровень заработка с учётом всех остальных факторов.

Hamermesh D.S., Biddle J.E. (1994) Beauty and the Labor Market, American Economic Review, 84, 1174–1194.

Данные:

```python
raw = pd.read_csv("beauty.csv", sep=";", index_col=False) 
raw.head()
```

Посмотрим на матрицу диаграмм рассеяния по количественным признакам:

```python
pd.plotting.scatter_matrix(raw[['wage', 'exper', 'educ', 'looks']], alpha=0.2,  
                                 figsize=(15, 15), diagonal='hist')
pylab.show()
# аргументы: данные (срез), прозрачность, размер графиков, тип диагональных
```

Видно что распределение уровня заработной платы скошено, с этим нужно что то сделать.

Оценим сбалансированность выборки по категориальным признакам:

```python
print raw.union.value_counts() # членство в профсоюзе
print raw.goodhlth.value_counts() # здоровье
print raw.black.value_counts() # цвет кожи
print raw.female.value_counts() # пол
print raw.married.value_counts() # семейное положение 
print raw.service.value_counts() # заслуги
```

У каждого признака все значения встречаются достаточно много раз, так что всё в порядке.

## Предобработка

```python
data = raw
```

Посмотрим на распределение целевого признака — уровня заработной платы: выборка не является нормальной, она скошена, посмотрим на логарифм данного признака

```python
plt.figure(figsize(16,7))

plt.subplot(121)
data['wage'].plot.hist(color='green')
plt.xlabel('Wage', fontsize=14)

# логорифмированная шкала
plt.subplot(122)
np.log(data['wage']).plot.hist(color='red')
plt.xlabel('Log wage', fontsize=14)

pylab.show()
```

Один человек в выборке получает 77.72\$ в час, остальные — меньше 45\$; удалим этого человека, чтобы регрессия на него не перенастроилась.

```python
data = data[data['wage'] < 77]
```

Посмотрим на распределение оценок привлекательности: 

```python
plt.figure(figsize(8,7))
# сортировка столбца Looks по существующим аргументам (карманам), по этим данным строим гистограмму
data.groupby('looks')['looks'].agg(lambda x: len(x)).plot(kind='bar', width=0.9)
plt.xticks(rotation=0)
plt.xlabel('Looks', fontsize=14)
pylab.show()
```

В группах looks=1 и looks=5 слишком мало наблюдений. Превратим признак looks в бинарный и закодируем с помощью двух фиктивных переменных:

```python
# apply формирует столбец по функции внутри нее
data['belowavg'] = data['looks'].apply(lambda x : 1 if x < 3 else 0) # ниже среднего
data['aboveavg'] = data['looks'].apply(lambda x : 1 if x > 3 else 0) # выше среднего
data.drop('looks', axis=1, inplace=True) # удаляем столбец looks
```

Данные теперь:

```python
data.head()
```

## Построение модели

### Простейшая модель

Построим линейную модель по всем признакам. Аргументы: отклик ~ признак_1 + признак_2 +...

```python
m1 = smf.ols('wage ~ exper + union + goodhlth + black + female + married +'\
                    'service + educ + belowavg + aboveavg', 
             data=data)
fitted = m1.fit()
print fitted.summary()
```

- R-squared: - R-квадрат показывает уровень того сколько мы можем объяснить дисперсии исходного признака (в данном случае 26%)
- Prob (F-statistic): - уровень значимости при гипотезе о том, что все признаки можно занулить
- Далее таблица:
- t - значение статистики Стьюдента при гипотезе о том что коэффициент признака равен нулю
- P>|t| - его уровень значимости, [0.025 0.975] - доверительный интервал

Посмотрим на распределение остатков:

```python
plt.figure(figsize(16,7))
plt.subplot(131)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab) # Q-Q график для ошибок модели (fitted.resid)

plt.subplot(132)
fitted.resid.plot.hist() # гисторграмма ошибок модели (fitted.resid)
plt.xlabel('Residuals', fontsize=14)

plt.subplot(133)
np.log(fitted.resid).plot.hist() # гисторграмма логарифма ошибок модели (fitted.resid)
plt.xlabel('Log(Residuals)', fontsize=14)
pylab.show()
```

Оно скошенное, как и исходный признак wage. В таких ситуациях часто помогает перейти от регрессии исходного признака к регрессии его логарифма.

### Логарифмируем отклик

```python
m2 = smf.ols('np.log(wage) ~ exper + union + goodhlth + black + female + married +'\
                            'service + educ + belowavg + aboveavg', data=data)
fitted = m2.fit()
print fitted.summary()

plt.figure(figsize(16,7))
plt.subplot(121)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)

plt.subplot(122)
fitted.resid.plot.hist()
plt.xlabel('Residuals', fontsize=14)
pylab.show()
```

Теперь стало лучше. Посмотрим теперь на зависимость остатков от непрерывных признаков (проверка на линейность отклика):

```python
plt.figure(figsize(16,7))

# образование от ошибки
plt.subplot(121)
scatter(data['educ'],fitted.resid)
plt.xlabel('Education', fontsize=14)
plt.ylabel('Residuals', fontsize=14)

# опыт работы от ошибки
plt.subplot(122)
scatter(data['exper'],fitted.resid)
plt.xlabel('Experience', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
pylab.show()
```

На втором графике видна квадратичная зависимость остатков от опыта работы. Попробуем добавить к признакам квадрат опыта работы, чтобы учесть этот эффект.

### Добавляем квадрат опыта работы

```python
m3 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + goodhlth + black + female +'\
                            'married + service + educ + belowavg + aboveavg', data=data)
fitted = m3.fit()
print fitted.summary()

plt.figure(figsize(16,7)) # Q-Q график
plt.subplot(121)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)

plt.subplot(122) # гистограмма ошибок
fitted.resid.plot.hist()
plt.xlabel('Residuals', fontsize=14)
plt.figure(figsize(16,5))

plt.subplot(131) # график образования от ошибки
scatter(data['educ'],fitted.resid)
plt.xlabel('Education', fontsize=14)
plt.ylabel('Residuals', fontsize=14)

plt.subplot(132) # график опыта работы от ошибки
scatter(data['exper'],fitted.resid)
plt.xlabel('Experience', fontsize=14)
plt.ylabel('Residuals', fontsize=14)

plt.subplot(133) # график квадрата опыта работы от ошибки
scatter(data['exper']**2,fitted.resid)
plt.xlabel('Experience^2', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
pylab.show()
```

Теперь в остатках нет зависимости от опыта работы

Используем критерий Бройша-Пагана для проверки гомоскедастичности ошибок:

```python
print 'Breusch-Pagan test: p=%f' % sms.het_breushpagan(fitted.resid, fitted.model.exog)[1]
# fitted.model.exog - экзогенные переменные модели
```

Ошибки гетероскедастичны, значит, значимость признаков может определяться неверно. Сделаем поправку Уайта (она применяется при случае гетероскедастичности ошибок, однако она менее точна, но другого выхода у нас нет): в итоге изменятся только оцнеки дисперсий наших коэффициентов регрессии

```python
m4 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + goodhlth + black + female +'\
                            'married + service + educ + belowavg + aboveavg', data=data)
fitted = m4.fit(cov_type='HC1') # будем испольтзовать первую поправку Уайта - HC1 (всего их где то 4)
print fitted.summary()

plt.figure(figsize(16,7))
plt.subplot(121)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)
plt.subplot(122)
fitted.resid.plot.hist()
plt.xlabel('Residuals', fontsize=14)
pylab.show()
```

### Удаляем незначимые признаки

В предыдущей модели незначимы (p-value>0.05): цвет кожи, здоровье, семейное положение. Удалим их. Индикатор привлекательности выше среднего тоже незначим, но удалять его не будем, потому что это одна из переменных, по которым на нужно в конце ответить на вопрос.

```python
m5 = smf.ols('np.log(wage) ~ exper + np.power(exper,2) + union + female + service + educ +'\
                            'belowavg + aboveavg', data=data)
fitted = m5.fit(cov_type='HC1')
print fitted.summary()

plt.figure(figsize(16,7))
plt.subplot(121)
sc.stats.probplot(fitted.resid, dist="norm", plot=pylab)
plt.subplot(122)
fitted.resid.plot.hist()
plt.xlabel('Residuals', fontsize=14)
pylab.show()
```

Посмотрим, не стала ли модель от удаления трёх признаков значимо хуже, с помощью критерия Фишера:

```python
print "F=%f, p=%f, k1=%f" % m4.fit().compare_f_test(m5.fit())
# Применим к обучению m4-модель большая, метод критерия Фишера на m5-модель меньшая (вложена в первую большую)
```

Не стала (Нулевая гипотеза о том, что удаленные признаки незначимы не отвергается)

Проверим, нет ли наблюдений, которые слишком сильно влияют на регрессионное уравнение (есть ли выбросы):

```python
plt.figure(figsize(8,7))
plot_leverage_resid2(fitted) # график ошибок и влиятельности точек
pylab.show()
# по OX-квадрат ошибки точки, OY-влиятельность точки (вычисляется посредством удаления каждой точки и замера отклонений)
```

```python
# точка на которой самая большая ошибка (у него очень большая зарплата, видимо мы предсказываем ему более маленькую зарплату)
data.loc[[269]]
```

Данные точки нельзя назвать выбросами, они не особо отклонены от основной масссы, мы их трогать не будем

## Выводы

Итоговая модель объясняет 40% вариации логарифма отклика. 

```python
# график прогноза экспоненты наших предсказаний (мы прогнозировали логорифм) от истиной заработной платы
plt.figure(figsize(16,7))
plt.subplot(121)
scatter(data['wage'],np.exp(fitted.fittedvalues))
plt.xlabel('Wage', fontsize=14)
plt.ylabel('Exponentiated predictions', fontsize=14)
plt.xlim([0,50])

# график прогноза наших предсказаний (мы прогнозировали логорифм) от логарифма истиной заработной платы
plt.subplot(122)
scatter(np.log(data['wage']),fitted.fittedvalues)
plt.xlabel('Log wage', fontsize=14)
plt.ylabel('Predictions', fontsize=14)
plt.xlim([0,4])
pylab.show()
```

При интересующих нас факторах привлекательности стоят коэффициенты -0.1307 (ниже среднего) и -0.0010 (выше среднего). 

Поскольку регрессия делалась на логарифм отклика, интерпретировать их можно как прирост в процентах. С учётом дополнительных факторов представители генеральной совокупности, из которой взята выборка, получают в среднем:

* на 13% меньше, если их привлекательность ниже среднего (p=0.001, 95% доверительный интервал — [5,21]%);
* столько же как и средний, если их привлекательность выше среднего (p=0.972, 95% доверительный интервал — [-6,6]%) - p-value большое, а интервал содержит 0, следовательно признак абсолютно неинформативен
